```python
def save_salary_rate():
    starting_annual_salary = float(input("your annual salary: "))
    total_cost = 1000000
    portion_down_payment = .25
    payable = total_cost * portion_down_payment
    
    semi_annual_raise = .07
    current_savings = 0
    r = .04
    monthly_salary = starting_annual_salary / 12    
    
    low = 0.0
    high = 1.0
    guess = (low + high) / 2.0
    num_guesses = 0
    flag = False
    
    while(abs(current_savings - payable) > 100):
        if(guess == 1.0):
            flag = True
            break
        
        num_guesses += 1
        print("number of guess:", num_guesses)
        print("low: ", low)
        print("high: ", high)
        print("new guess: ", guess)
        
        
        
        annual_salary = starting_annual_salary
        monthly_salary = annual_salary/12
        month = 0
        months = 36
        current_savings = 0
        
        while( month < months):
            
            current_savings += (current_savings * r /12) + (guess * monthly_salary)
            month += 1
            if ((month % 6) == 0):
                annual_salary += annual_salary * semi_annual_raise
                monthly_salary = annual_salary / 12
                
        if(abs(current_savings - (payable)) > 100):
            print("current_savings: ",current_savings)
            if(current_savings < (payable)):
                low = guess
            else:
                high = guess
                
            guess = (low + high) / 2.0
            
            
    if(flag):
        print("It is not possible to pay down payment in 3 years.")
    else:         
        print("Best savings rate: ", guess)
        print("Steps in bisection search: ", num_guesses)
        
        
                
        
    
save_salary_rate()
```
